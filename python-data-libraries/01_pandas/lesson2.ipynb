```python
import pandas as pd
```

```python
data = {
    'Employee': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Department': ['Data Science', 'Power BI', 'ML Engineering',
                   'Data Science', 'ML Engineering'],
    'Salary': [95000, 75000, 110000, 98000, 115000],
    'Experience_Yrs': [4, 2, 6, 5, 7]
}
df = pd.DataFrame(data)
df
```

```python
# -- Column Selection ---
print("=== Singel Column (Series) ===")
df['Salary']
```

```python
print("==== Multiple Columns (DataFrame) ===")
df[['Employee', 'Department']]
```

```python
# --- Row Filtering ---
print("=== Salary > 100000 ====")
df[df['Salary']>100000]
```

```python
print("=== Data Science AND Experience > 3 ===")
df[(df['Department'] == 'Data Science') & (df['Experience_Yrs']>3)]
```

```python
print("=== NOT Pwer BI ===")
df[~(df['Department'] == 'Power BI')]
```

```python
# -- loc & iloc ---
print("=== loc: rows 0-2, Employee & Salary ===")
df.loc[0:2, ['Employee', 'Salary']]
```

```python
print("=== iloc: first 3 rows, first 2 columns ===")
df.iloc[0:3, [0, 1]]
```
