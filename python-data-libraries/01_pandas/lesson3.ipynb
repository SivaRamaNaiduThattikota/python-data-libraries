```python
import pandas as pd
import numpy as np  # needed to create NaN values

# Messy dataset — like real world data!
data = {
    'Employee':   ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Bob'],
    'Department': ['Data Science', 'ML Engineering', 'ML Engineering', 
                   'data science', None, 'Power BI'],
    'Salary':     [95000, 75000, None, 98000, 115000, 75000],
    'Experience': [4, 2, 6, None, 7, 2]
}
df = pd.DataFrame(data)
print("=== ORIGINAL MESSY DATA ===")
print(df)

# --- STEP 1: Check missing values ---
print("\n=== MISSING VALUES ===")
print(df.isnull().sum())

# --- STEP 2: Fill missing Salary with mean ---
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print("\n=== After filling Salary with mean ===")
print(df['Salary'])

# --- STEP 3: Fill missing Department with 'Unknown' ---
df['Department'] = df['Department'].fillna('Unknown')

# --- STEP 4: Drop rows where Experience is missing ---
df.dropna(subset=['Experience'], inplace=True)

# --- STEP 5: Fix Department — strip spaces & fix casing ---
df['Department'] = df['Department'].str.strip()
df['Department'] = df['Department'].str.title()  # Title Case
print("\n=== After fixing Department ===")
print(df['Department'])

# --- STEP 6: Remove duplicate rows ---
print("\n=== Duplicates found:", df.duplicated().sum(), "===")
df.drop_duplicates(inplace=True)

# --- STEP 7: Rename column ---
df.rename(columns={'Experience': 'Experience_Yrs'}, inplace=True)

# --- FINAL CLEAN DATA ---
print("\n=== FINAL CLEAN DATA ===")
print(df)
print("\nShape:", df.shape)
print("\nMissing values left:", df.isnull().sum().sum())
```
