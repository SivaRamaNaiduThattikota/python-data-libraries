```python
# Step 1 - Import pandas
import pandas as pd
```

```python
# Step 2 - Create a DataFrame
data = {
    'Drug': ['Humira', 'Skyrizi', 'Rinvoq', 'Botox', 'Vraylar'],
    'Revenue_B': [14.3, 4.1, 2.8, 2.6, 1.9],
    'Year': [2023, 2023, 2023, 2023, 2023],
    'Therapy_Area':['Immunology','Immunology','Immunology','Neuroscience','Neuroscience']
}

df = pd.DataFrame(data)
df
```

```python
# Step 3 - Explore it
print("=== HEAD ===")
df.head()
```

```python
print("=== SHAPE ===")
df.shape
```

```python
print("=== INFO ===")
df.info()
```

```python
print("=== DESCRIBE ===")
df.describe()
```
