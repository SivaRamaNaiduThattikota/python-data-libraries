<!-- ATTRIBUTE CARD -->
<div class="attr-card">
  <div class="card-title">🔹 <span class="attr-label">Attribute</span> — It's a stored property (like a label)</div>
  <ul class="bullet-list">
    <li>It just holds a value — no action needed</li>
    <li>Like asking <em>"how tall are you?"</em> — the answer is already there</li>
    <li>No parentheses because you're just reading a value</li>
  </ul>
  <div class="code-block">
    <span class="attr">df.shape</span>  &nbsp;&nbsp;&nbsp; <span class="comment"># just returns (5, 4) — already stored</span><br>

    <span class="attr">df.columns</span> &nbsp;&nbsp; <span class="comment"># just returns column names — already stored</span><br>
    <span class="attr">df.size</span>  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="comment"># just returns total cells — already stored</span>
  </div>
</div>

<!-- METHOD CARD -->
<div class="method-card">
  <div class="card-title">🔸 <span class="method-label">Method</span> — It's an action/function (does something)</div>
  <ul class="bullet-list">
    <li>It performs an operation and returns a result</li>
    <li>Like asking <em>"can you summarize this?"</em> — it has to do work</li>
    <li>Needs parentheses because it's a function call</li>
  </ul>
  <div class="code-block">
    <span class="method">df.info()</span>  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="comment"># performs action → prints full summary</span><br>
    <span class="method">df.describe()</span> &nbsp;&nbsp; <span class="comment"># performs action → calculates statistics</span><br>
    <span class="method">df.head(3)</span> &nbsp;&nbsp;&nbsp;&nbsp; <span class="comment"># performs action → slices first 3 rows</span><br>
    <span class="method">df.tail()</span>  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="comment"># performs action → slices last 5 rows</span>

<table class="styled">
  <tr><th>Operator</th><th>Symbol</th><th>Example</th></tr>
  <tr><td>AND</td><td><code>&amp;</code></td><td>Both conditions must be true</td></tr>
  <tr><td>OR</td><td><code>|</code></td><td>Either condition must be true</td></tr>
  <tr><td>NOT</td><td><code>~</code></td><td>Opposite of condition</td></tr>

  df[(df['Department'] == 'Data Science') | (df['Department'] == 'ML Engineering')]<br><br>

  df[~(df['Department'] == 'Power BI')]

<div class="part-header"><h2>PART 3 — loc vs iloc</h2></div>
<div class="info-box">👇 This confuses many people — let's make it simple!</div>

<table class="styled">
  <tr><th></th><th>loc[]</th><th>iloc[]</th></tr>
  <tr><td><strong>Full Name</strong></td><td>Label locator</td><td>Integer locator</td></tr>
  <tr><td><strong>Uses</strong></td><td>Column/row names</td><td>Column/row numbers</td></tr>
  <tr><td><strong>Think of it as</strong></td><td>Search by name</td><td>Search by position</td></tr>
</table>

<h3 class="sub">🔹 loc[] — by Labels (names)</h3>
<div class="code-block"><span style="color:#6c7086;"># Get row 0 to 2, columns 'Employee' and 'Salary'</span><br>df.loc[0:2, ['Employee', 'Salary']]</div>
<div class="output-block">&nbsp; Employee &nbsp; Salary<br>0 &nbsp;&nbsp; Alice &nbsp;&nbsp; 95000<br>1 &nbsp;&nbsp;&nbsp; Bob &nbsp;&nbsp; 75000<br>2 &nbsp;Charlie &nbsp; 110000</div>

<h3 class="sub">🔹 iloc[] — by Position (numbers)</h3>
<div class="code-block"><span style="color:#6c7086;"># Get rows 0 to 2, columns at position 0 and 2</span><br>df.iloc[0:3, [0, 2]]</div>
<div class="output-block">&nbsp; Employee &nbsp; Salary<br>0 &nbsp;&nbsp; Alice &nbsp;&nbsp; 95000<br>1 &nbsp;&nbsp;&nbsp; Bob &nbsp;&nbsp; 75000<br>2 &nbsp;Charlie &nbsp; 110000</div>

<h3 class="sub">💡 Easy Way to Remember</h3>
<div class="memory-box">
  <span class="loc">loc</span> &nbsp; → Letters &nbsp; → Labels/Names<br>
  <span class="iloc">iloc</span> → Integer &nbsp; → Index numbers (positions)
</div>

<h3 class="sub">Edge Case 1 — Filtering with NaN values</h3>
<div class="code-block">
  data = {<br>
  &nbsp;&nbsp;&nbsp;&nbsp;<span class="c-string">'Employee'</span>: [<span class="c-string">'Alice'</span>, <span class="c-string">'Bob'</span>, <span class="c-string">'Charlie'</span>],<br>
  &nbsp;&nbsp;&nbsp;&nbsp;<span class="c-string">'Salary'</span>: [<span class="c-number">95000</span>, <span class="c-keyword">None</span>, <span class="c-number">110000</span>]<br>
  }<br>
  df = pd.<span class="c-method">DataFrame</span>(data)<br><br>
  df[df[<span class="c-string">'Salary'</span>] > <span class="c-number">90000</span>]
</div>
<div class="output-block">
  &nbsp;&nbsp; Employee &nbsp;&nbsp;&nbsp; Salary<br>
  0 &nbsp;&nbsp;&nbsp; Alice &nbsp;&nbsp; 95000.0<br>
  2 &nbsp; Charlie &nbsp; 110000.0
</div>
<div class="warn-box">
  ⚠️ <strong>Bob is silently excluded</strong> — NaN comparisons always return <code style="background:#3a2a00;padding:2px 6px;border-radius:4px;color:#f9e2af;">False</code>!<br>
  <code style="background:#3a2a00;padding:2px 6px;border-radius:4px;color:#f9e2af;">NaN > 90000 = False</code> → row just disappears without any warning!
</div>
<div class="code-block">
  <span class="c-comment"># ✅ Always handle NaN BEFORE filtering!</span><br>
  df.<span class="c-method">dropna</span>(subset=[<span class="c-string">'Salary'</span>])<br><br>
  <span class="c-comment"># OR filter safely with notna()</span><br>
  df[df[<span class="c-string">'Salary'</span>].<span class="c-method">notna</span>() &amp; (df[<span class="c-string">'Salary'</span>] > <span class="c-number">90000</span>)]
</div>

<hr class="divider">

<!-- Edge Case 2 -->
<h3 class="sub">Edge Case 2 — loc vs iloc with reset_index</h3>
<div class="code-block">
  df2 = df.<span class="c-method">drop</span>(index=<span class="c-number">1</span>) &nbsp;&nbsp; <span class="c-comment"># drop Bob</span><br>
  <span class="c-method">print</span>(df2)
</div>
<div class="output-block">
  &nbsp;&nbsp; Employee &nbsp;&nbsp;&nbsp; Salary<br>
  0 &nbsp;&nbsp;&nbsp; Alice &nbsp;&nbsp; 95000.0<br>
  <span style="color:#f38ba8;">2 &nbsp; Charlie &nbsp; 110000.0 &nbsp;&nbsp; ← index jumps from 0 to 2!</span>
</div>
<div class="code-block">
  df2.<span class="c-method">iloc</span>[<span class="c-number">1</span>] &nbsp;&nbsp; <span class="c-comment"># ✅ returns Charlie (position 1)</span><br>
  df2.<span class="c-method">loc</span>[<span class="c-number">1</span>] &nbsp;&nbsp;&nbsp; <span class="c-comment"># ❌ KeyError! index label 1 doesn't exist anymore!</span>
</div>
<div class="danger-box">
  ❌ <strong>KeyError</strong> — index label <code style="background:#2e1e1e;padding:2px 6px;border-radius:4px;color:#f38ba8;">1</code> was dropped with Bob! <code style="background:#2e1e1e;padding:2px 6px;border-radius:4px;color:#f38ba8;">loc</code> uses labels, not positions.
</div>
<div class="code-block">
  <span class="c-comment"># ✅ Always reset_index after dropping rows!</span><br>
  df2 = df2.<span class="c-method">reset_index</span>(drop=<span class="c-keyword">True</span>)<br>
  df2.<span class="c-method">loc</span>[<span class="c-number">1</span>] &nbsp;&nbsp; <span class="c-comment"># ✅ now works! Returns Charlie</span>
</div>

<hr class="divider">

<!-- Edge Case 3 -->
<h3 class="sub">Edge Case 3 — Chained filtering pitfall</h3>
<div class="code-block">
  df[df[<span class="c-string">'Salary'</span>] > <span class="c-number">90000</span>][<span class="c-string">'Department'</span>] = <span class="c-string">'Updated'</span><br>
  <span class="c-comment"># ⚠️ SettingWithCopyWarning — change may NOT apply to original df!</span>
</div>
<div class="warn-box">
  ⚠️ <strong>SettingWithCopyWarning</strong> — chaining <code style="background:#3a2a00;padding:2px 6px;border-radius:4px;color:#f9e2af;">[][]</code> creates a copy, not a view. Your change silently disappears!
</div>
<div class="code-block">
  <span class="c-comment"># ✅ Always use loc for filtering + assigning!</span><br>
  df.<span class="c-method">loc</span>[df[<span class="c-string">'Salary'</span>] > <span class="c-number">90000</span>, <span class="c-string">'Department'</span>] = <span class="c-string">'Updated'</span>
</div>
<div class="note-box">✅ <strong>Rule:</strong> filter + assign = always <code style="background:#313244;padding:2px 6px;border-radius:4px;color:#a6e3a1;">df.loc[condition, 'column'] = value</code></div>

<hr class="divider">

<!-- Edge Case 4 -->
<h3 class="sub">Edge Case 4 — & and | operator precedence</h3>
<div class="code-block">
  df[df[<span class="c-string">'Salary'</span>] > <span class="c-number">90000</span> &amp; df[<span class="c-string">'Experience'</span>] > <span class="c-number">3</span>]<br>
  <span class="c-comment"># ❌ ERROR or wrong result — & binds tighter than &gt;!</span>
</div>
<div class="danger-box">
  ❌ <code style="background:#2e1e1e;padding:2px 6px;border-radius:4px;color:#f38ba8;">&amp;</code> has higher precedence than <code style="background:#2e1e1e;padding:2px 6px;border-radius:4px;color:#f38ba8;">&gt;</code> — Python evaluates <code style="background:#2e1e1e;padding:2px 6px;border-radius:4px;color:#f38ba8;">90000 &amp; df['Experience']</code> first, giving wrong results or an error!
</div>
<div class="code-block">
  <span class="c-comment"># ✅ Always wrap EVERY condition in parentheses!</span><br>
  df[(df[<span class="c-string">'Salary'</span>] > <span class="c-number">90000</span>) &amp; (df[<span class="c-string">'Experience'</span>] > <span class="c-number">3</span>)]
</div>

<hr class="divider">

<!-- ── GOLDEN RULES ── -->
<div class="golden-header"><h2>💡 Golden Rules — Lesson 2</h2></div>

<div class="golden-box">
  <span class="step-num">1.</span> &nbsp;Always wrap every condition &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="step-method">( )</span> &nbsp;around each when using <span class="step-method">&amp;</span> or <span class="step-method">|</span><br>
  <span class="step-num">2.</span> &nbsp;Handle NaN BEFORE filtering &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="step-method">dropna()</span> or <span class="step-method">notna()</span> &nbsp;<span class="step-note">— NaN comparisons = False silently!</span><br>
  <span class="step-num">3.</span> &nbsp;Single bracket &nbsp;<span class="step-method">df['col']</span> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; → Series &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span class="step-note">| Double bracket </span><span class="step-method">df[['col']]</span><span class="step-note"> → DataFrame</span><br>
  <span class="step-num">4.</span> &nbsp;<span class="step-method">loc</span> uses LABELS &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="step-note">breaks if index has gaps after dropping rows</span><br>
  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span class="step-method">iloc</span> uses POSITIONS &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="step-note">always safe regardless of index</span><br>
  <span class="step-num">5.</span> &nbsp;Filter + assign &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="step-method">df.loc[condition, 'col'] = value</span> &nbsp;<span class="step-note">— never chain [][]!</span><br>
  <span class="step-num">6.</span> &nbsp;After dropping rows &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="step-method">reset_index(drop=True)</span> &nbsp;<span class="step-note">— before using loc!</span>
</div>

<div class="trap-header"><h2>🪤 Common Traps — Lesson 2</h2></div>

<!-- Trap 1 -->
<h3 class="sub">🪤 Trap 1 — Using = instead of == in filter</h3>
<div class="code-block">
  df[df[<span class="c-string">'Department'</span>] = <span class="c-string">'Data Science'</span>]  &nbsp; <span class="c-comment"># ❌ SyntaxError!</span><br>
  df[df[<span class="c-string">'Department'</span>] == <span class="c-string">'Data Science'</span>] &nbsp; <span class="c-comment"># ✅</span>
</div>
<div class="danger-box">❌ <code style="background:#2e1e1e;padding:2px 6px;border-radius:4px;color:#f38ba8;">=</code> is assignment &nbsp;·&nbsp; <code style="background:#2e1e1e;padding:2px 6px;border-radius:4px;color:#f38ba8;">==</code> is comparison — always use <code style="background:#2e1e1e;padding:2px 6px;border-radius:4px;color:#f38ba8;">==</code> inside filters!</div>

<!-- Trap 2 -->
<h3 class="sub">🪤 Trap 2 — Using 'and' instead of '&'</h3>
<div class="code-block">
  df[df[<span class="c-string">'Salary'</span>] > <span class="c-number">90000</span> <span class="c-keyword">and</span> df[<span class="c-string">'Experience'</span>] > <span class="c-number">3</span>] &nbsp; <span class="c-comment"># ❌ ValueError!</span><br>
  df[(df[<span class="c-string">'Salary'</span>] > <span class="c-number">90000</span>) &amp; (df[<span class="c-string">'Experience'</span>] > <span class="c-number">3</span>)] &nbsp; <span class="c-comment"># ✅</span>
</div>
<div class="danger-box">❌ <code style="background:#2e1e1e;padding:2px 6px;border-radius:4px;color:#f38ba8;">and</code> / <code style="background:#2e1e1e;padding:2px 6px;border-radius:4px;color:#f38ba8;">or</code> are Python keywords — use <code style="background:#2e1e1e;padding:2px 6px;border-radius:4px;color:#f38ba8;">&amp;</code> and <code style="background:#2e1e1e;padding:2px 6px;border-radius:4px;color:#f38ba8;">|</code> for pandas filtering!</div>

<!-- Trap 3 -->
<h3 class="sub">🪤 Trap 3 — Using 'or' instead of '|'</h3>
<div class="code-block">
  df[df[<span class="c-string">'Salary'</span>] > <span class="c-number">90000</span> <span class="c-keyword">or</span> df[<span class="c-string">'Experience'</span>] > <span class="c-number">3</span>]  &nbsp; <span class="c-comment"># ❌ ValueError!</span><br>
  df[(df[<span class="c-string">'Salary'</span>] > <span class="c-number">90000</span>) | (df[<span class="c-string">'Experience'</span>] > <span class="c-number">3</span>)] &nbsp; <span class="c-comment"># ✅</span>
</div>

<!-- Trap 4 -->
<h3 class="sub">🪤 Trap 4 — loc with non-existent label after dropping</h3>
<div class="code-block">
  df2 = df.<span class="c-method">drop</span>(index=<span class="c-number">1</span>)<br>
  df2.<span class="c-method">loc</span>[<span class="c-number">1</span>] &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># ❌ KeyError — index 1 was dropped!</span><br>
  df2.<span class="c-method">reset_index</span>(drop=<span class="c-keyword">True</span>, inplace=<span class="c-keyword">True</span>)<br>
  df2.<span class="c-method">loc</span>[<span class="c-number">1</span>] &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># ✅ works now!</span>
</div>
<div class="note-box">✅ <strong>Always</strong> <code style="background:#313244;padding:2px 6px;border-radius:4px;color:#a6e3a1;">reset_index(drop=True)</code> after dropping rows — then <code style="background:#313244;padding:2px 6px;border-radius:4px;color:#a6e3a1;">loc</code> is safe again!</div>

<!-- ── PART 1 ── -->
<div class="part-header"><h2>PART 1 — Handling Missing Values (NaN)</h2></div>

<h3 class="sub">🔹 What is NaN?</h3>
<div class="note-box">
  <strong>NaN</strong> = Not a Number = missing / empty value<br>
  Like an <strong>empty cell in Excel!</strong>
</div>

<h3 class="sub">🔹 Detect Missing Values</h3>
<div class="code-block">
  df.<span class="c-method">isnull</span>() &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># True/False for every cell</span><br>
  df.<span class="c-method">isnull</span>().<span class="c-method">sum</span>() &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># Count of missing values per column</span><br>
  df.<span class="c-method">isnull</span>().<span class="c-method">sum</span>().<span class="c-method">sum</span>() &nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># Total missing values in entire df</span>
</div>

<h3 class="sub">🔹 Drop Missing Values — <span class="tag-red">dropna()</span></h3>
<div class="code-block">
  df.<span class="c-method">dropna</span>() &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># Drop rows with ANY missing value</span><br>
  df.<span class="c-method">dropna</span>(subset=[<span class="c-string">'Salary'</span>]) &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># Drop rows where Salary is missing only</span><br>
  df.<span class="c-method">dropna</span>(axis=<span class="c-number">1</span>) &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># Drop COLUMNS with missing values</span>
</div>

<h3 class="sub">🔹 Fill Missing Values — <span class="tag">fillna()</span></h3>
<div class="code-block">
  df.<span class="c-method">fillna</span>(<span class="c-number">0</span>) &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># Fill ALL missing with 0</span><br>
  df[<span class="c-string">'Salary'</span>].<span class="c-method">fillna</span>(<span class="c-number">0</span>) &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># Fill missing Salary with 0</span><br>
  df[<span class="c-string">'Salary'</span>].<span class="c-method">fillna</span>(df[<span class="c-string">'Salary'</span>].<span class="c-method">mean</span>()) &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># Fill with average salary</span><br>
  df[<span class="c-string">'Department'</span>].<span class="c-method">fillna</span>(<span class="c-string">'Unknown'</span>) &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># Fill text column with 'Unknown'</span>
</div>

<h3 class="sub">💡 dropna vs fillna — When to Use Which?</h3>
<table class="styled">
  <tr><th>Situation</th><th>Use</th></tr>
  <tr><td>Very few missing rows, data is large</td><td><span class="tag-red">dropna()</span> — just remove them</td></tr>
  <tr><td>Important column, can't lose rows</td><td><span class="tag">fillna(mean/median)</span> — fill smartly</td></tr>
  <tr><td>Text / category column missing</td><td><span class="tag">fillna('Unknown')</span></td></tr>
  <tr><td>Too many missing in a column</td><td><span class="tag-red">drop()</span> — remove entire column</td></tr>
</table>

<div class="part-header"><h2>PART 2 — Renaming Columns — rename()</h2></div>

<div class="code-block">
  <span class="c-comment"># Rename specific columns</span><br>
  df.<span class="c-method">rename</span>(columns={<br>
  &nbsp;&nbsp;&nbsp;&nbsp;<span class="c-string">'Experience_Yrs'</span> : <span class="c-string">'Experience'</span>,<br>
  &nbsp;&nbsp;&nbsp;&nbsp;<span class="c-string">'Salary'</span> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; : <span class="c-string">'Annual_Salary'</span><br>
  }, inplace=<span class="c-keyword">True</span>)
</div>

<div class="warn-box">
  ⚠️ <strong>inplace=True</strong> → modifies the original DataFrame directly<br>
  Without it → returns a new DataFrame but <strong>doesn't change the original!</strong>
</div>

<h3 class="sub">💡 inplace=True explained</h3>
<div class="code-block">
  <span class="c-comment"># Without inplace → original df unchanged!</span><br>
  df.<span class="c-method">rename</span>(columns={<span class="c-string">'Salary'</span>: <span class="c-string">'Annual_Salary'</span>}) &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># ❌ df still has 'Salary'</span><br><br>
  <span class="c-comment"># With inplace → original df updated!</span><br>
  df.<span class="c-method">rename</span>(columns={<span class="c-string">'Salary'</span>: <span class="c-string">'Annual_Salary'</span>}, inplace=<span class="c-keyword">True</span>) &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># ✅</span><br><br>
  <span class="c-comment"># OR assign back to df</span><br>
  df = df.<span class="c-method">rename</span>(columns={<span class="c-string">'Salary'</span>: <span class="c-string">'Annual_Salary'</span>}) &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># ✅ also works!</span>
</div>

<!-- ── PART 4 ── -->

  df = df.<span class="c-method">replace</span>(<span class="c-string">''</span>,  np.nan) &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># convert '' to NaN first</span><br>

  <span class="s-check">✅</span> &nbsp;<span class="s-method">str.strip()</span> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="s-arrow">→</span> &nbsp;<span class="s-desc">remove spaces &nbsp;</span><span class="s-must">⚠️ MUST assign back!</span><br>

  <span class="s-check">✅</span> &nbsp;<span class="s-method">str.title()</span> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="s-arrow">→</span> &nbsp;<span class="s-desc">fix casing &nbsp;</span><span class="s-must">⚠️ MUST assign back!</span><br>

  <hr class="divider">

  <span class="s-check">✅</span> &nbsp;<span class="s-method">drop_duplicates()</span> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="s-arrow">→</span> &nbsp;<span class="s-desc">remove truly identical rows</span><br>

  <span class="s-check">✅</span> &nbsp;<span class="s-method">rename(columns={})</span> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="s-arrow">→</span> &nbsp;<span class="s-desc">rename columns</span><br>

  <hr class="divider">

  <span class="s-check">✅</span> &nbsp;<span class="s-method">dropna()</span> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="s-arrow">→</span> &nbsp;<span class="s-desc">only removes </span><span style="color:#a6e3a1;">NaN / None</span> &nbsp;<span class="s-warn">— NOT &nbsp;'' &nbsp;0 &nbsp;False!</span>

</div>

<!-- ── TITLE ── -->
<div class="part-header"><h2>🔍 How drop_duplicates() Really Works</h2></div>

<!-- ── CORRECT THOUGHT ── -->
<h3 class="sub">✅ Your First Thought — CORRECT!</h3>
<div class="correct-card">
  <span style="color:#6c7086;">Row 1:</span> &nbsp;<span class="c-same">Bob</span> | <span class="c-same">ML Engineering</span> | <span class="c-same">75000</span> | <span class="c-same">2</span><br>
  <span style="color:#6c7086;">Row 5:</span> &nbsp;<span class="c-same">Bob</span> | <span class="c-same">ML Engineering</span> | <span class="c-same">75000</span> | <span class="c-same">2</span><br>
  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-arrow">↑</span> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span class="c-arrow">↑</span> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span class="c-arrow">↑</span> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span class="c-arrow">↑</span><br>
  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span style="color:#f9e2af;">SAME</span> &nbsp;&nbsp;&nbsp; <span style="color:#f9e2af;">SAME</span> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span style="color:#f9e2af;">SAME</span> &nbsp;&nbsp; <span style="color:#f9e2af;">SAME</span> &nbsp; <span class="c-arrow">→</span> <span class="c-bad">DUPLICATE! Drop one!</span>
</div>
<div class="note-box">✅ It compares <strong>entire rows</strong> — ALL columns must match to be a duplicate!</div>

<hr class="divider">

<!-- ── WRONG THOUGHT ── -->
<h3 class="sub">❌ Your Second Thought — NOT How it Works!</h3>
<div class="wrong-card">
  <span style="color:#6c7086;">Row 1:</span> &nbsp;Alice | <span class="c-same">Data Science</span> | 95000 | 4<br>
  <span style="color:#6c7086;">Row 2:</span> &nbsp;Bob &nbsp;&nbsp;| <span class="c-same">Data Science</span> | 75000 | 2<br>
  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-arrow">↑</span><br>
  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span style="color:#f9e2af;">SAME department</span> — <span class="c-bad">but NOT a duplicate!</span>
</div>
<div class="danger-box">❌ Just because ONE column matches doesn't mean the row is dropped! By default — <strong>ALL columns must be identical!</strong></div>

<hr class="divider">

<!-- ── SUBSET ── -->
<div class="part-header"><h2>🎯 BUT — You Can Check Specific Columns with subset=</h2></div>

<div class="code-block">
  <span class="c-comment"># Default — ALL columns must match</span><br>
  df.<span class="c-method">drop_duplicates</span>()<br><br>
  <span class="c-comment"># Only check 'Employee' column — if name repeats, drop it!</span><br>
  df.<span class="c-method">drop_duplicates</span>(subset=[<span class="c-string">'Employee'</span>])<br><br>
  <span class="c-comment"># Only check Employee + Department combination</span><br>
  df.<span class="c-method">drop_duplicates</span>(subset=[<span class="c-string">'Employee'</span>, <span class="c-string">'Department'</span>])
</div>

<h3 class="sub">📊 Example</h3>
<div class="visual-box">
  <span style="color:#6c7086;">Row 0:</span> &nbsp;Alice | Data Science &nbsp;&nbsp; | 95000 | 4<br>
  <span style="color:#6c7086;">Row 1:</span> &nbsp;Bob &nbsp;&nbsp;| ML Engineering | 75000 | 2<br>
  <span style="color:#6c7086;">Row 2:</span> &nbsp;<span class="c-same">Bob</span> &nbsp;&nbsp;| Power BI &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; | 80000 | 3 &nbsp;&nbsp; <span style="color:#f38ba8;">← Bob again!</span><br><br>
  <span class="c-comment"># Default drop_duplicates() → keeps ALL 3 (no full row match)</span><br><br>
  <span class="c-comment"># drop_duplicates(subset=['Employee'])</span><br>
  <span class="c-arrow">→</span> Drops Row 2 because <span class="c-string">'Bob'</span> already appeared in Row 1!
</div>

<hr class="divider">

<!-- ── KEEP PARAMETER ── -->
<div class="part-header"><h2>🗂️ Which Row Gets Deleted? — keep Parameter</h2></div>

<div class="code-block">
  <span class="c-comment"># Default — keep FIRST occurrence, drop rest</span><br>
  df.<span class="c-method">drop_duplicates</span>(keep=<span class="c-string">'first'</span>) &nbsp;&nbsp; <span class="c-comment"># ← default behavior</span><br><br>
  <span class="c-comment"># Keep LAST occurrence, drop first</span><br>
  df.<span class="c-method">drop_duplicates</span>(keep=<span class="c-string">'last'</span>)<br><br>
  <span class="c-comment"># Drop ALL occurrences — even the first one!</span><br>
  df.<span class="c-method">drop_duplicates</span>(keep=<span class="c-keyword">False</span>)
</div>

<h3 class="sub">👁️ Visual Example</h3>
<div class="visual-box">
  <span style="color:#6c7086;">Row 0:</span> &nbsp;<span class="c-same">Bob | ML Engineering | 75000 | 2</span><br>
  <span style="color:#6c7086;">Row 1:</span> &nbsp;<span class="c-same">Bob | ML Engineering | 75000 | 2</span> &nbsp;&nbsp; <span style="color:#f38ba8;">← duplicate</span><br>
  <span style="color:#6c7086;">Row 2:</span> &nbsp;<span class="c-same">Bob | ML Engineering | 75000 | 2</span> &nbsp;&nbsp; <span style="color:#f38ba8;">← duplicate</span><br><br>
  keep=<span class="c-string">'first'</span> &nbsp; <span class="c-arrow">→</span> &nbsp;Keeps Row 0, drops Row 1 &amp; 2 &nbsp;&nbsp; <span class="badge-keep">✅ keep Row 0</span><br>
  keep=<span class="c-string">'last'</span> &nbsp;&nbsp; <span class="c-arrow">→</span> &nbsp;Keeps Row 2, drops Row 0 &amp; 1 &nbsp;&nbsp; <span class="badge-keep">✅ keep Row 2</span><br>
  keep=<span class="c-keyword">False</span> &nbsp;&nbsp;&nbsp; <span class="c-arrow">→</span> &nbsp;Drops ALL 3 rows! &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="badge-warn">⚠️ all gone!</span>
</div>

<!-- ── EDGE CASES ── -->
<div class="warn-header"><h2>⚠️ Edge Cases — The Tricky Ones!</h2></div>

<!-- Edge Case 1 -->
<h3 class="sub">Edge Case 1 — NaN values in duplicates</h3>
<div class="code-block">
  data = {<br>
  &nbsp;&nbsp;&nbsp;&nbsp;<span class="c-string">'Employee'</span>: [<span class="c-string">'Bob'</span>, <span class="c-string">'Bob'</span>],<br>
  &nbsp;&nbsp;&nbsp;&nbsp;<span class="c-string">'Salary'</span>: [<span class="c-keyword">None</span>, <span class="c-keyword">None</span>] &nbsp;&nbsp; <span class="c-comment"># both NaN</span><br>
  }
</div>
<div class="warn-box">❗ Two NaN values ARE considered equal by <code style="background:#3a2a00;padding:2px 5px;border-radius:4px;color:#f9e2af;">drop_duplicates()</code> — this row WILL be flagged as duplicate and dropped!</div>
<div class="code-block">
  df.<span class="c-method">drop_duplicates</span>()<br>
  <span class="c-comment"># Result: keeps only 1 Bob row ✅</span>
</div>

<!-- Edge Case 2 -->
<h3 class="sub">Edge Case 2 — Space difference</h3>
<div class="code-block">
  data = {<br>
  &nbsp;&nbsp;&nbsp;&nbsp;<span class="c-string">'Employee'</span>: [<span class="c-string">'Bob'</span>, <span class="c-string">'Bob'</span>],<br>
  &nbsp;&nbsp;&nbsp;&nbsp;<span class="c-string">'Department'</span>: [<span class="c-string">'ML Engineering'</span>, <span class="c-string">' ML Engineering'</span>] &nbsp;<span class="c-comment"># space!</span><br>
  }
</div>
<div class="danger-box">❗ <code style="background:#2e1e1e;padding:2px 5px;border-radius:4px;color:#f38ba8;">'ML Engineering'</code> ≠ <code style="background:#2e1e1e;padding:2px 5px;border-radius:4px;color:#f38ba8;">' ML Engineering'</code> — Always clean strings BEFORE dropping duplicates!</div>
<div class="code-block">
  <span class="c-comment"># ✅ Correct order always:</span><br>
  df[<span class="c-string">'Department'</span>] = df[<span class="c-string">'Department'</span>].str.<span class="c-method">strip</span>() &nbsp;&nbsp; <span class="c-comment"># clean first!</span><br>
  df.<span class="c-method">drop_duplicates</span>(inplace=<span class="c-keyword">True</span>) &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># then drop!</span>
</div>

<!-- Edge Case 3 -->
<h3 class="sub">Edge Case 3 — Case difference</h3>
<div class="code-block">
  data = {<br>
  &nbsp;&nbsp;&nbsp;&nbsp;<span class="c-string">'Employee'</span>: [<span class="c-string">'Bob'</span>, <span class="c-string">'bob'</span>], &nbsp;&nbsp; <span class="c-comment"># different case!</span><br>
  &nbsp;&nbsp;&nbsp;&nbsp;<span class="c-string">'Salary'</span>: [<span class="c-number">75000</span>, <span class="c-number">75000</span>]<br>
  }
</div>
<div class="danger-box">❗ <code style="background:#2e1e1e;padding:2px 5px;border-radius:4px;color:#f38ba8;">'Bob'</code> ≠ <code style="background:#2e1e1e;padding:2px 5px;border-radius:4px;color:#f38ba8;">'bob'</code> to pandas — NOT considered duplicate! Always fix casing BEFORE dropping duplicates!</div>
<div class="code-block">
  <span class="c-comment"># ✅ Correct order:</span><br>
  df[<span class="c-string">'Employee'</span>] = df[<span class="c-string">'Employee'</span>].str.<span class="c-method">lower</span>() &nbsp;&nbsp; <span class="c-comment"># lowercase first!</span><br>
  df.<span class="c-method">drop_duplicates</span>(inplace=<span class="c-keyword">True</span>) &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># then drop!</span>
</div>

<!-- Edge Case 4 -->
<h3 class="sub">Edge Case 4 — Data type difference</h3>
<div class="code-block">
  data = {<br>
  &nbsp;&nbsp;&nbsp;&nbsp;<span class="c-string">'Year'</span>: [<span class="c-number">2023</span>, <span class="c-string">'2023'</span>] &nbsp;&nbsp; <span class="c-comment"># int vs string!</span><br>
  }
</div>
<div class="danger-box">❗ <code style="background:#2e1e1e;padding:2px 5px;border-radius:4px;color:#f38ba8;">2023 (int)</code> ≠ <code style="background:#2e1e1e;padding:2px 5px;border-radius:4px;color:#f38ba8;">'2023' (string)</code> — NOT duplicate! Always fix data types BEFORE dropping duplicates!</div>
<div class="code-block">
  <span class="c-comment"># ✅ Correct order:</span><br>
  df[<span class="c-string">'Year'</span>] = df[<span class="c-string">'Year'</span>].<span class="c-method">astype</span>(<span class="c-keyword">int</span>) &nbsp;&nbsp; <span class="c-comment"># fix type first!</span><br>
  df.<span class="c-method">drop_duplicates</span>(inplace=<span class="c-keyword">True</span>) &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-comment"># then drop!</span>
</div>

<!-- Edge Case 5 -->
<h3 class="sub">Edge Case 5 — Index doesn't reset after dropping!</h3>
<div class="code-block">
  df.<span class="c-method">drop_duplicates</span>(inplace=<span class="c-keyword">True</span>)<br>
  <span class="c-method">print</span>(df)
</div>
<div class="output-block">
  &nbsp;&nbsp; Employee &nbsp; Salary<br>
  0 &nbsp;&nbsp;&nbsp; Alice &nbsp;&nbsp; 95000<br>
  <span style="color:#f38ba8;">2 &nbsp; Charlie &nbsp; 110000 &nbsp;&nbsp; ← index jumps! 1 is missing</span><br>
  <span style="color:#f38ba8;">4 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Eve &nbsp; 115000 &nbsp;&nbsp; ← index jumps! 3 is missing</span>
</div>
<div class="danger-box">❗ Index has gaps after dropping rows — this can cause bugs later in ML! Always reset index after dropping!</div>
<div class="code-block">
  <span class="c-comment"># ✅ Reset index after dropping</span><br>
  df.<span class="c-method">drop_duplicates</span>(inplace=<span class="c-keyword">True</span>)<br>
  df.<span class="c-method">reset_index</span>(drop=<span class="c-keyword">True</span>, inplace=<span class="c-keyword">True</span>) &nbsp;&nbsp; <span class="c-comment"># drop=True removes old index!</span>
</div>
<div class="output-block">
  &nbsp;&nbsp; Employee &nbsp; Salary<br>
  <span style="color:#a6e3a1;">0 &nbsp;&nbsp;&nbsp; Alice &nbsp;&nbsp; 95000 &nbsp;&nbsp; ← clean index!</span><br>
  <span style="color:#a6e3a1;">1 &nbsp; Charlie &nbsp; 110000 &nbsp;&nbsp; ← clean index!</span><br>
  <span style="color:#a6e3a1;">2 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Eve &nbsp; 115000 &nbsp;&nbsp; ← clean index!</span>
</div>

<div class="golden-box">
  <span class="step-num">1.</span> &nbsp;Fix data types &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="c-arrow">→</span> &nbsp;<span class="step-method">astype()</span><br>
