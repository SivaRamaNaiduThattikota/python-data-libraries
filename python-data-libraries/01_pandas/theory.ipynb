<!-- SIMPLY EXPLAINED -->
<div class="part-header"><h3>🔍 Attribute vs Method — Simply Explained</h3></div>

<!-- ATTRIBUTE CARD -->
<div class="attr-card">
  <div class="card-title">🔹 <span class="attr-label">Attribute</span> — It's a stored property (like a label)</div>
  <ul class="bullet-list">
    <li>It just holds a value — no action needed</li>
    <li>Like asking <em>"how tall are you?"</em> — the answer is already there</li>
    <li>No parentheses because you're just reading a value</li>
  </ul>
  <div class="code-block">
    <span class="attr">df.shape</span>  &nbsp;&nbsp;&nbsp; <span class="comment"># just returns (5, 4) — already stored</span><br>
    <span class="attr">df.dtypes</span> &nbsp;&nbsp;&nbsp; <span class="comment"># just returns column types — already stored</span><br>
    <span class="attr">df.columns</span> &nbsp;&nbsp; <span class="comment"># just returns column names — already stored</span><br>
    <span class="attr">df.size</span>  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="comment"># just returns total cells — already stored</span>
  </div>
</div>

<!-- METHOD CARD -->
<div class="method-card">
  <div class="card-title">🔸 <span class="method-label">Method</span> — It's an action/function (does something)</div>
  <ul class="bullet-list">
    <li>It performs an operation and returns a result</li>
    <li>Like asking <em>"can you summarize this?"</em> — it has to do work</li>
    <li>Needs parentheses because it's a function call</li>
  </ul>
  <div class="code-block">
    <span class="method">df.info()</span>  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="comment"># performs action → prints full summary</span><br>
    <span class="method">df.describe()</span> &nbsp;&nbsp; <span class="comment"># performs action → calculates statistics</span><br>
    <span class="method">df.head(3)</span> &nbsp;&nbsp;&nbsp;&nbsp; <span class="comment"># performs action → slices first 3 rows</span><br>
    <span class="method">df.tail()</span>  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <span class="comment"># performs action → slices last 5 rows</span>
  </div>
</div>

<h1 class="main-title">📊 Selecting & Filtering Data</h1>

<!-- THEORY -->
<div class="theory-box">
  <h2 style="color:#89b4fa;margin-top:0;font-family:'Segoe UI',sans-serif;">📖 THEORY — Why is This Important for ML?</h2>
  <p style="color:#cdd6f4;font-family:'Segoe UI',sans-serif;">In every ML project, your dataset has many columns and thousands of rows — you'll never use all of them. You need to:</p>
  <div class="bullet">
    <span class="check">✅</span> Select only the columns your model needs<br>
    <span class="check">✅</span> Filter only the rows that match certain conditions<br>
    <span class="check">✅</span> Locate specific data points
  </div>
  <div class="info-box" style="margin-top:14px;">💡 Think of it like <strong>Power BI filters & column selection</strong> — but in code!</div>
</div>

<hr class="divider">

<!-- PART 1 -->
<div class="part-header"><h2>PART 1 — Selecting Columns</h2></div>

<h3 class="sub">🔹 Select a Single Column → returns a Series</h3>
<div class="code-block">df['Salary']</div>
<div class="output-block">0     95000<br>1     75000<br>2    110000<br>3     98000<br>4    115000<br>Name: Salary, dtype: int64</div>
<div class="note-box">⚠️ Single column → returns <strong>Series</strong> (1D — like a single list)</div>

<h3 class="sub">🔹 Select Multiple Columns → returns a DataFrame</h3>
<div class="code-block">df[['Employee', 'Salary']]</div>
<div class="output-block">&nbsp; Employee &nbsp; Salary<br>0 &nbsp;&nbsp; Alice &nbsp;&nbsp; 95000<br>1 &nbsp;&nbsp;&nbsp; Bob &nbsp;&nbsp; 75000<br>2 &nbsp;Charlie &nbsp; 110000<br>3 &nbsp;&nbsp; Diana &nbsp;&nbsp; 98000<br>4 &nbsp;&nbsp;&nbsp; Eve &nbsp;&nbsp; 115000</div>
<div class="note-box">✅ Double brackets <code style="background:#313244;padding:2px 6px;border-radius:4px;color:#a6e3a1;">[[]]</code> → returns <strong>DataFrame</strong> (2D — like a table)</div>

<h3 class="sub">💡 Single vs Double Brackets — Common Confusion!</h3>
<table class="styled">
  <tr><th>Syntax</th><th>Returns</th><th>Think of it as</th></tr>
  <tr><td><code>df['Salary']</code></td><td>Series (1 column)</td><td>Single column from Excel</td></tr>
  <tr><td><code>df[['Salary']]</code></td><td>DataFrame (table)</td><td>Table with 1 column</td></tr>
  <tr><td><code>df[['Salary', 'Dept']]</code></td><td>DataFrame (table)</td><td>Table with 2 columns</td></tr>
</table>

<hr class="divider">

<!-- PART 2 -->
<div class="part-header"><h2>PART 2 — Filtering Rows (Conditions)</h2></div>

<h3 class="sub">🔹 Basic Filter — Single Condition</h3>
<div class="code-block"><span style="color:#6c7086;"># Rows where Salary > 100000</span><br>df[df['Salary'] > 100000]</div>
<div class="output-block">&nbsp; Employee &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Department &nbsp;&nbsp; Salary &nbsp; Experience_Yrs<br>2 &nbsp;Charlie &nbsp; ML Engineering &nbsp; 110000 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 6<br>4 &nbsp;&nbsp;&nbsp;&nbsp; Eve &nbsp; ML Engineering &nbsp; 115000 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 7</div>

<h3 class="sub">🔹 Multiple Conditions</h3>
<table class="styled">
  <tr><th>Operator</th><th>Symbol</th><th>Example</th></tr>
  <tr><td>AND</td><td><code>&amp;</code></td><td>Both conditions must be true</td></tr>
  <tr><td>OR</td><td><code>|</code></td><td>Either condition must be true</td></tr>
  <tr><td>NOT</td><td><code>~</code></td><td>Opposite of condition</td></tr>
</table>
<div class="code-block">
  <span style="color:#6c7086;"># AND — Salary > 90000 AND Experience > 4</span><br>
  df[(df['Salary'] > 90000) &amp; (df['Experience_Yrs'] > 4)]<br><br>
  <span style="color:#6c7086;"># OR — Department is Data Science OR ML Engineering</span><br>
  df[(df['Department'] == 'Data Science') | (df['Department'] == 'ML Engineering')]<br><br>
  <span style="color:#6c7086;"># NOT — Everyone except Power BI</span><br>
  df[~(df['Department'] == 'Power BI')]
</div>
<div class="warn-box">⚠️ <strong>Important:</strong> Always wrap each condition in <code style="background:#3a2a00;padding:2px 6px;border-radius:4px;color:#f9e2af;">()</code> when using <code style="background:#3a2a00;padding:2px 6px;border-radius:4px;color:#f9e2af;">&amp;</code> or <code style="background:#3a2a00;padding:2px 6px;border-radius:4px;color:#f9e2af;">|</code></div>

<hr class="divider">

<!-- PART 3 -->
<div class="part-header"><h2>PART 3 — loc vs iloc</h2></div>
<div class="info-box">👇 This confuses many people — let's make it simple!</div>

<table class="styled">
  <tr><th></th><th>loc[]</th><th>iloc[]</th></tr>
  <tr><td><strong>Full Name</strong></td><td>Label locator</td><td>Integer locator</td></tr>
  <tr><td><strong>Uses</strong></td><td>Column/row names</td><td>Column/row numbers</td></tr>
  <tr><td><strong>Think of it as</strong></td><td>Search by name</td><td>Search by position</td></tr>
</table>

<h3 class="sub">🔹 loc[] — by Labels (names)</h3>
<div class="code-block"><span style="color:#6c7086;"># Get row 0 to 2, columns 'Employee' and 'Salary'</span><br>df.loc[0:2, ['Employee', 'Salary']]</div>
<div class="output-block">&nbsp; Employee &nbsp; Salary<br>0 &nbsp;&nbsp; Alice &nbsp;&nbsp; 95000<br>1 &nbsp;&nbsp;&nbsp; Bob &nbsp;&nbsp; 75000<br>2 &nbsp;Charlie &nbsp; 110000</div>

<h3 class="sub">🔹 iloc[] — by Position (numbers)</h3>
<div class="code-block"><span style="color:#6c7086;"># Get rows 0 to 2, columns at position 0 and 2</span><br>df.iloc[0:3, [0, 2]]</div>
<div class="output-block">&nbsp; Employee &nbsp; Salary<br>0 &nbsp;&nbsp; Alice &nbsp;&nbsp; 95000<br>1 &nbsp;&nbsp;&nbsp; Bob &nbsp;&nbsp; 75000<br>2 &nbsp;Charlie &nbsp; 110000</div>

<h3 class="sub">💡 Easy Way to Remember</h3>
<div class="memory-box">
  <span class="loc">loc</span> &nbsp; → Letters &nbsp; → Labels/Names<br>
  <span class="iloc">iloc</span> → Integer &nbsp; → Index numbers (positions)
</div>
